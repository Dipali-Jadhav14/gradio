```python
import gradio as gr

# function to add two numbers
def add_numbers(a, b):
    return a + b

# Gradio interface
demo = gr.Interface(
    fn=add_numbers,
    inputs=[
        gr.Number(label="First Number"),
        gr.Number(label="Second Number")
    ],
    outputs=gr.Number(label="Sum"),
    title="Addition App",
    description="Enter two numbers and get their sum"
)

# launch app
demo.launch()













```

```python
import gradio as gr

# login function
def login(username, password):
    # dummy credentials
    if username == "admin" and password == "1234":
        return " Login successful!"
    else:
        return " Invalid username or password"

# Gradio UI
demo = gr.Interface(
    fn=login,
    inputs=[
        gr.Textbox(label="Username", placeholder="Enter username"),
        gr.Textbox(label="Password", type="password", placeholder="Enter password")
    ],
    outputs=gr.Textbox(label="Status"),
    title="Login Form",
    description="Simple login form using Gradio"
)

# launch app
demo.launch()









```

```python
import gradio as gr

# login logic
def login(username, password):
    if username == "user" and password == "1234":
        return "<h3 style='color:green;'> Login Successful</h3>"
    else:
        return "<h3 style='color:red;'>Invalid Username or Password</h3>"

# custom CSS
custom_css = """
body {
    background: linear-gradient(120deg, #89f7fe, #66a6ff);
}
.login-box {
    background: white;
    padding: 25px;
    border-radius: 10px;
    width: 320px;
    margin: auto;
    box-shadow: 0px 10px 25px rgba(0,0,0,0.2);
}
.login-box h2 {
    text-align: center;
}
"""

with gr.Blocks(css=custom_css) as demo:
    gr.HTML("<div class='login-box'><h2>User Login</h2></div>")

    username = gr.Textbox(label="Username", placeholder="Enter username")
    password = gr.Textbox(label="Password", type="password", placeholder="Enter password")

    login_btn = gr.Button("Login")
    output = gr.HTML()

    login_btn.click(
        fn=login,
        inputs=[username, password],
        outputs=output
    )

demo.launch()
```

```python
import gradio as gr

# temporary user storage
users = {}

# registration function
def register(username, password):
    if username in users:
        return "<p style='color:red;'> User already exists</p>"
    users[username] = password
    return "<p style='color:green;'> Registration successful</p>"

# login function
def login(username, password):
    if username in users and users[username] == password:
        return "<h3 style='color:green;'>ðŸŽ‰ Login Successful</h3>"
    return "<h3 style='color:red;'> Invalid credentials</h3>"

# custom CSS
custom_css = """
body {
    background: linear-gradient(135deg, #667eea, #764ba2);
    font-family: Arial, sans-serif;
}
.card {
    background: white;
    padding: 25px;
    border-radius: 12px;
    width: 360px;
    margin: auto;
    box-shadow: 0 15px 30px rgba(0,0,0,0.25);
}
.card h2 {
    text-align: center;
    color: #333;
}
"""

with gr.Blocks(css=custom_css) as demo:

    gr.HTML("<div class='card'><h2>User Registration</h2></div>")
    reg_user = gr.Textbox(label="Username")
    reg_pass = gr.Textbox(label="Password", type="password")
    reg_btn = gr.Button("Register")
    reg_out = gr.HTML()

    reg_btn.click(register, [reg_user, reg_pass], reg_out)

    gr.HTML("<hr>")

    gr.HTML("<div class='card'><h2>User Login</h2></div>")
    log_user = gr.Textbox(label="Username")
    log_pass = gr.Textbox(label="Password", type="password")
    log_btn = gr.Button("Login")
    log_out = gr.HTML()

    log_btn.click(login, [log_user, log_pass], log_out)

demo.launch()
```
